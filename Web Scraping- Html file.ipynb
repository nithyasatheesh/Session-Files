#### Extract data from Website :goodreads.com

```python
#Do a get request to the server where the webpage is hosted
```

```python
import requests
url="https://www.goodreads.com/quotes"
```

```python
html=requests.get(url).text # get the request (get() method sends a GET request to the specified url)
```

```python
requests.get(url).status_code
# returns a number that indicates the status (200 is OK, 404 is Not Found)
```

```python
#pulling data out of HTML and XML files and parse.
soup=BeautifulSoup(html,"html.parser")
```

```python
# extract title of the page
soup.title.text
```

```python
# extract author name
soup.find_all('span',class_="authorOrTitle")
```

```python
soup.find_all('span',class_="authorOrTitle")[0].text.strip()
```

```python
#Authors
authors=[]
for i in soup.find_all('span',class_="authorOrTitle"):
    authors.append(i.text.strip())
```

```python
print(authors)
```

```python
soup.find_all('a',class_="smallText")[0].text.split()[0]
```

```python
#Number of Likes
likes=[]
for i in soup.find_all('a',class_="smallText"):
    likes.append(i.text.split()[0])
```

```python
print(likes)
```

```python
#Quotes
quotes=[]
for i in soup.find_all('div',class_="quoteText"):
    quotes.append(list(i.children)[0].strip())
```

```python
#alternative
soup.find_all('div',class_="quoteText")[0].text.split('\n')[1].strip()
```

```python
#list(soup.find_all('div',class_="quoteText")[0].children)[0].strip()
```

```python
quotes
```

```python
import pandas as pd
table=pd.DataFrame({'Author':authors,'Quotes':quotes,'Likes':likes})
table.shape
```

```python
table.head()
```

```python
#Function get soup and returns table
def get_data(soup):
    #Authors
    authors=[]
    for i in soup.find_all('span',class_="authorOrTitle"):
        authors.append(i.text.strip())
    likes=[]
    for i in soup.find_all('a',class_="smallText"):
        likes.append(i.text.split()[0])
    quotes=[]
    for i in soup.find_all('div',class_="quoteText"):
        quotes.append(list(i.children)[0].strip())
    table=pd.DataFrame({'Author':authors,'Quotes':quotes,'Likes':likes})
    return table
```

```python
data=get_data(soup)
data.shape
```

```python
data.head(2)
```

```python
stub="https://www.goodreads.com/quotes?page={}"
```

```python
stub.format(1)
```

```python
from tqdm import tqdm
q=[]
for i in tqdm(range(1,30)):
    url=stub.format(i)
    html=requests.get(url).text
    soup=BeautifulSoup(html,'html.parser')
    q.append(get_data(soup))    
```

```python
q
```

```python
df=pd.concat(q,axis=0)
df
```

```python
def get_table(soup):
    #Author
    author=[]
    for i in soup.find_all("a",class_="authorName"):
        author.append(i.span.text.strip())
    #Average rating
    ratings=[]
    for i in soup.find_all("span",class_='minirating'):
        ratings.append(i.text.split()[0])
    name=[]
    for i in soup.find_all("a",class_="bookTitle"):
        name.append(i.span.text.strip())
    table=pd.DataFrame({'name':name,'author':author,'ratings':ratings})
    return table
```

```python
stub="https://www.goodreads.com/list/show/19.Best_for_Book_Clubs?page={}"
stub.format(20)
```

```python
T=[]
for i in range(1,10):
    url=stub.format(i)
    html=requests.get(url).text
    soup=BeautifulSoup(html,'html.parser')
    T.append(get_table(soup))
```

```python
df=pd.concat(T,axis=0)
df.head()
```

```python
state=[]
confirmed=[]
active=[]
recovered=[]
deaths=[]
iterable=iter(covid['statewise'])
next(iterable)
for i in iterable:
    state.append(i['state'])
    confirmed.append(i['confirmed'])
    active.append(i['active'])
    recovered.append(i['recovered'])
    deaths.append(i['deaths'])
```
